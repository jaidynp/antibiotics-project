# Setup

```python
# Import libraries
import pandas as pd
import numpy as np
import random

# Agglomerative clustering library
from sklearn.cluster import AgglomerativeClustering
```

```python
# Load dataset
df = pd.read_csv('small_GNPS_edges.tsv', sep='\t')

# View dataset
df
```

# Agglomerative Clustering

```python
# Create precomputed matrix - this will be used as input for the 'fit' method.

# 'distance' is an np.array of floats with values from the Cosine distance column.
distance = df['Cosine'].to_numpy().astype(np.float64)

# Reshape the data.
distance = distance.reshape(-1, 1)
distance
```

```python
# Transform 'distance' np.array into a distance matrix.
from scipy.spatial import distance_matrix
distance_matrix = pd.DataFrame(distance_matrix(distance, distance), index=df.index, columns=df.index)
distance_matrix
```

```python
# Create model
# Parameters: affinity = 'precomputed', linkage = {'complete', 'average', 'single'}
# Note from sklearn:  If affinity =  “precomputed”, a distance matrix (instead of a similarity matrix) is needed as input for the fit method.
cluster = AgglomerativeClustering(affinity = 'precomputed', linkage = 'average')

# Fit model using distance matrix
cluster.fit(distance_matrix)
```

```python
# Output how the data points have been clustered. 
cluster.labels_
```

# Ward Heirarchial Clustering

```python
# Create model
# Parameters: affinity = 'euclidean', linkage = 'ward'
# Note from sklearn:  If linkage is “ward”, only affinity = “euclidean” is accepted.
ward_cluster = AgglomerativeClustering(affinity = 'euclidean', linkage = 'ward')

# Fit model using Cos distance column
ward_cluster.fit(distance) 
```

# Partitioning Around Medoids (rpy2)

```python
# Documentation
# https://rpy2.github.io/doc/latest/html/index.html
# https://astrostatistics.psu.edu/su07/R/html/cluster/html/pam.html
# https://www.datanovia.com/en/lessons/k-medoids-in-r-algorithm-and-practical-examples/
# https://www.rdocumentation.org/packages/cluster/versions/2.1.0/topics/pam

# Import libraries
import rpy2
import rpy2.robjects as robjects               # We utilize a 'high-level' interface over a 'low-level' interface. 
from rpy2.robjects.packages import importr     # Import R packages. 
import rpy2.robjects.packages as rpackages

# Pandas conversion of rpy2
from rpy2.robjects import pandas2ri
pandas2ri.activate()
```

```python
# Import R's 'base' and 'utils' packages.
base = rpackages.importr('base')
utils = rpackages.importr('utils')

# Select a mirror for R packages
utils.chooseCRANmirror(ind=56) # select CRAN mirror 

# Import cluster package
r_cluster = rpackages.importr('cluster')
```

```python
# Load and view data as an r object
data = robjects.r('read.table(file = "small_GNPS_edges.tsv", header = T)')
data
```

```python
# Generate random integer for k
k = random.randint(2, 10)

# Create model
model = r_cluster.pam(df['Cosine'], k)

# View model results
print(k)
print(model)
```

# Erdos Renyi using python-igraph

```python
# Documentation
# https://igraph.org/python/doc/igraph.GraphBase-class.html#Erdos_Renyi
# https://igraph.org/python/doc/tutorial/tutorial.html

# Import libraries
import igraph
from igraph import *
```

```python
# Generate random values for number of nodes (int n) and edges (float p, sampled uniformly between 0-1)
n = random.randint(1, 50)                   # number of nodes
p = np.random.uniform(low = 0, high = 1)    # probability for edge creation


# Create graph using random values
# n = number of nodes, p = probability for edge creation, undirected
g = igraph.Graph.Erdos_Renyi(n, p, directed=False, loops=False)

# View graph
print('N = '+ str(n), 'P = '+ str(p))
plot(g)
```

# Erdos Renyi using networkx

```python
# Documentation:
# https://www.geeksforgeeks.org/erdos-renyl-model-generating-random-graphs/
# https://networkx.github.io/documentation/networkx-1.9/reference/generated/networkx.generators.random_graphs.erdos_renyi_graph.html

# Import libraries
import networkx as nx               # Generate erdos renyi graph
import matplotlib.pyplot as plt     # View graph
```

```python
# Generate random values for number of nodes (int n) and edges (float p, sampled uniformly between 0-1)
n = random.randint(1, 50)                   # number of nodes
p = np.random.uniform(low = 0, high = 1)    # probability for edge creation

# Create graph using random values
# n = number of nodes, p = probability for edge creation, undirected
G = nx.erdos_renyi_graph(n, p, directed = False) 
nx.draw(G, with_labels=True) 

a = 'N = '+ str(n), 'P = '+ str(p)
plt.title(a)
plt.show()
```
